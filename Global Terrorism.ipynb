## Task 4 - Exploratory Data Analysis on Global Terrorism

```python
#Importing the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading the dataset
df=pd.read_csv("global_terrorism.csv",encoding='latin',low_memory=False)
```

```python
#reading the first 5 rows
df.head()
```

```python
df.info
```

```python
df.describe()
```

```python
colors = ['red', 'green', 'yellow', 'blue', 'black', 'pink', 'orange', 'cyan', 'violet', 'purple']
df.columns
```

## Top 5 Regions with the most attacks

```python
plt.figure(figsize=[15,8])
df['region_txt'].value_counts()[:5].plot.bar(color=colors);
plt.title("Total Number of attacks Region Wise")
plt.xlabel("Regions")
plt.ylabel('Number of attacks')
plt.show()
```

## Total Number of Attacks Country Wise

```python

plt.figure(figsize=[15,8])
df['country_txt'].value_counts()[:10].plot.bar(color=colors);
plt.title("Total Number of attacks Country Wise")
plt.xlabel("Countries")
plt.ylabel('Number of attacks')
plt.show()
```

## The 5 most common types of attacks

```python
attack_c = df["attacktype1_txt"].value_counts()[:6]
attacks = ['Bombing/Explosion','Armed Assault', 'Assasination', 'Hostage Taking (Kidnapping)', 'Facility/Infrastructure Attack', 'Unknown'];
fig = plt.figure(figsize =(10, 10)) 
plt.pie(attack_c, labels = attacks, autopct='%1.1f%%')
plt.show()
```

## Top 5 Weapons used in Attacks

```python
weapon_c = df['weaptype1_txt'].value_counts()[:5]
weapon = ["Explosives", "Firearms", "Unknown", "Incendiary", "Melee"]
fig = plt.figure(figsize =(10, 10)) 
plt.pie(weapon_c, labels = weapon,autopct='%1.1f%%')
plt.show()
```

## Analysis on the outcome of the attacks 

```python
attack_counts = df["success"].value_counts()
fig = plt.figure(figsize =(10, 10)) 
plt.pie(attack_counts, labels = ["success", "failed"], autopct='%1.1f%%')
plt.show()
```

## Analysis on the outcome of the attacks based on the type of attack

```python
attack_count = df["attacktype1_txt"].value_counts()[:5]
attack_count
```

```python
attack_successc = df.groupby("attacktype1_txt")["success"].sum().sort_values(ascending=False)[:5]
attack_successc
```

```python
totalkeys = list(attack_count)
successkeys = list(attack_successc)
total_attacks = {"TotalAttacks": totalkeys, "SuccessAttacks": successkeys}
```

## Comparing the total number of attacks with it's successful number of attacks

```python
attacks = attack_count.index
dataFrame = pd.DataFrame(data=total_attacks, index=attacks)
dataFrame.plot.bar()
```

## Success rate of the attacks based on the type of attack

```python
success_rate = (attack_successc/attack_count)*100
fig = plt.figure(figsize =(10, 10)) 
plt.pie(success_rate, labels = ["Bombing/Explosion", "Armed Assault", "Assasination", "Hostage Taking", 'Facility/Infrastructure Attack'], autopct='%1.1f%%' )
plt.show()
```

## Year wise terrorism

```python
plt.subplots(figsize=(20,10))
sns.countplot('iyear',data=df,palette='rainbow')
plt.xticks(rotation=50)
plt.xlabel("Year")
plt.ylabel("Number of Attacks")
plt.title('Number of Attacks Year Wise')
plt.show()
```
